```python
def age_group_creation(row):
    if row["age"] < 30 :
        return "20 - 30"
    elif row["age"] < 40:
        return "30 - 40"
    elif row["age"] < 50:
        return "40 - 50"
    elif row["age"] < 60:
        return "50 - 60"
    else:
        return "60 - 70"
```

```python
data["age group"] = data.apply(age_group_creation, axis= 1)
```

```python
data
```

```python
data["age group"].value_counts()
```

```python
data.groupby("age group")["annual Salary"].mean().sort_values(ascending=False).plot(kind= "bar", color = "Orange")
plt.title("Age Group and Annual Salary")
plt.xlabel("Age Group")
plt.ylabel("Annual Salary")
plt.show()
```

```python
data.groupby("age group")["car purchase amount"].mean().sort_values(ascending=False).plot(kind = "bar", color = "Cyan")
plt.title("Car Purchase amount by age group")
plt.ylabel("Car Purchase Amount")
plt.show()
```

```python
X = data[["age", "annual Salary", "net worth"]]
y = data[["car purchase amount"]]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)
```

```python
len(X_test), len(X_train)
```

```python
len(y_test), len(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python

X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
def modelresult(predictions):
    print("Mean absolute error on model is {} ".format(mean_absolute_error(y_test, predictions)))
    print("Root Mean Squared error on model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr = lr.fit(X_train, y_train) 
```

```python
prediction_lr= lr.predict(X_test)
```

```python
modelresult(prediction_lr)
```

```python
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C" : [0.01, 0.1, 0.5], "kernel" : ["linear", "rbf", "poly"], "degree" : [2, 3, 4]}
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train.values.ravel())
```

```python
prediction_gridsvr= gridsvr.predict(X_test)
```

```python
modelresult(prediction_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
param_gridrfr = {"max_depth" : [8, 9, 10, 11, 12], "n_estimators" : [11, 12, 10, 13, 14, 15, 16, 17]}
```

```python
prediction_rfr = gridrfr.predict(X_test)
```

```python
modelresult(prediction_rfr)
```
